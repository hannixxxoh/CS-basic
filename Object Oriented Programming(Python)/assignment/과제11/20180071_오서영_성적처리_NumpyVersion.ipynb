```python
import numpy as np
import matplotlib.pyplot as plt

Student = np.array([[75, 89, 92, 78],
               [90, 76, 88, 83],
               [55, 99, 81, 88],
               [80, 67, 75, 91]])
subname = ['kor', 'Math', 'Eng', 'Com']
stdname = ['Lee', 'Park', 'Jung', 'Choi']

submean = []
stdmean = []
submax = []


totalmean = np.mean(np.sum(Student, axis=0))
print("총점의 반 평균", totalmean)

print('\n각 과목의 평균')
for i in range(4):
    temp = np.mean(Student[:, i])
    print(subname[i], temp, end=' ')
    submean.append(temp)
    
print('\n\n각 학생별 평균')
for i in range(4):
    temp = np.mean(Student[i, :])
    print(stdname[i], temp, end=' ')
    stdmean.append(temp)

print('\n\n각 과목의 최고점')
for i in range(4):
    temp = np.max(Student[:, i])
    print(subname[i], temp, end=' ')
    submax.append(temp)
    
plt.plot(submean, label='subject mean')
plt.plot(submax, label='subject max')
plt.xticks([0, 1, 2, 3], labels=subname)
plt.legend()
plt.show()

plt.plot(stdmean, label='student mean')
plt.legend()
plt.xticks([0, 1, 2, 3], labels=stdname)
plt.show()
```

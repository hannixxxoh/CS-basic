```python
import numpy as np
import matplotlib.pyplot as plt

with plt.xkcd():
    X = np.linspace(-3, 3, 256)
    C = np.cos(X)
    plt.plot(X, C)
    plt.scatter([2*np.pi/3], [-0.5], color='r', linewidth=5)
    #plt.plot(2*np.pi/3, -0.5, c='red', lw=1, marker='o', ms=10)
    plt.title('XKCD style plot!!!')
    plt.annotate(r'-0.5 HERE!', xy=(2*np.pi/3, -0.5), xytext=(-100, -50), xycoords='data',
                 textcoords='offset points', arrowprops=dict(arrowstyle="->", linewidth=1))
    plt.xticks([-3, -2, -1, 0, 1, 2, 3])
```
